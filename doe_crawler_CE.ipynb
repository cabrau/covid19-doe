```python
#http://imagens.seplag.ce.gov.br/PDF/20200115/do20200115p01.pdf
#ano-mes-dia-p01
base_url = 'http://imagens.seplag.ce.gov.br/PDF/2020'
months = ['01','02','03','04'] #,'05','06','07','08','09', '10', '11', '12']
days = ['01','02','03','04','05','06','07','08','09']
parts = ['p01','p02','p03','p04','p05','p06','p07','p08','p09',]
ce_path = 'DOEs/CE/'
downloaded = []

for file in os.listdir(ce_path):
    downloaded.append(file)

for d in range(10,32):
    days.append(str(d))
    
for m in months:
    for d in days:
        for p in parts:
            do = 'do2020'+m+d+p+'.pdf'
            if do in downloaded:
                print(do + ' raspado')
                continue
            url = base_url+m+d+'/'+do
            myfile = requests.get(url)
            
            
            if myfile.status_code == 200:
                print(url + ' baixado')                
                file_path = ce_path + do
                open(file_path, 'wb').write(myfile.content)
    


#url = 'http://imagens.seplag.ce.gov.br/PDF/20200106/do20200106p05.pdf'
#myfile = requests.get(url)
#open('/home/cabral/Documents/notebooks/covid19_DOE_crawler/DOEs/CE/do20200106p01.pdf', 'wb').write(myfile.content)
```
