```python
import PyPDF2
import pandas as pd
import os
import re
```

```python
def parseParagraphs(text,word_list):
    text = text.lower()
    word_list = [x.lower() for x in word_list]
    word_set = set(word_list)
    text_list = text.split('*** *** ***')
    paragraphs = []
    for paragraph in text_list:
        paragraph_set = set(paragraph.split())
        if len(word_set.intersection(paragraph_set)) > 0:
            paragraph = paragraph.replace('\n',' ')
            paragraphs.append(paragraph)
            
    return paragraphs

def getDate(do):
    do = do.replace('.pdf','')
    do = do.replace('do','')
    date = do.split('p')[0]
    year = date[0:4]
    month = date[4:6]
    day = date[6:]
    date = day+'/'+month+'/'+year
    return date

def getURL(base,do):
    #CE http://imagens.seplag.ce.gov.br/PDF/20200106/do20200106p05.pdf
    date = do.replace('.pdf','')
    date = date.replace('do','')
    date = date.split('p')[0]
    url = base+date+do
    return url
```

```python
with open(r'does_raspados.txt') as file:
    files_scrapped = file.read()
files_scrapped = files_scrapped.split('\n')[0:-1]
files_scrapped
```

```python
path = '/home/cabral/Documents/notebooks/covid19_DOE_crawler/DOEs'
state = 'CE'

keywords = ['coronav√≠rus','coronavirus','corona virus', 
            'covid-19','covid19','covid 19', 'coe 2019-ncov']

bases = {'CE': 'http://imagens.seplag.ce.gov.br/PDF/'}

paragraphs = []
state = []
DOE = []
urls = []
dates = []

#search in all directories
for directorie in os.listdir(path):
    print(directorie)
    dir_path = path+'/'+directorie
    
    #acess every file in every directory
    for file in os.listdir(dir_path):
        
        if file in files_scrapped:
            continue
            
        print(file)
        file_path = dir_path+'/'+file
        pdfFileObj = open(file_path, 'rb')
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        num_pages = pdfReader.numPages
        text = ''
        url = getURL(bases[directorie],file)
        date = getDate(file)
        
        #iterate over every page in file
        for i in range(num_pages):
            pageObj = pdfReader.getPage(i)
            try:
                file_text = pageObj.extractText()
            except Exception as inst:
                file_text = ''
                print(inst)
            text += file_text
            
        p = parseParagraphs(text,keywords)
        s = [directorie] * len(p)
        f = [file] * len(p)
        u = [url] * len(p)
        d = [date] * len(p)
        
        if len(p) > 0:
            files_scrapped.append(file)
        
        
        paragraphs.extend(p)
        state.extend(s)
        DOE.extend(f)
        urls.extend(u)
        dates.extend(d)    
```

```python
df = pd.DataFrame({'TEXTO':paragraphs,'ESTADO': state, 'DATA':dates, 'ARQUIVO': DOE, 'URL': urls})
df.head(10)
```

```python
df['TEXTO'] = [str(msg) for msg in list(df['TEXTO'])] 
df['TEXTO'] = [msg.strip() for msg in list(df['TEXTO'])]
df['TEXTO'] = [re.sub('\s+',' ',msg) for msg in list(df['TEXTO'])]
df.head(10)
```

```python
df_old = pd.read_csv('DOE_COVID19.csv')
frames = [df_old, df]
result = pd.concat(frames)
result = result.drop_duplicates(subset = 'TEXTO')
```
